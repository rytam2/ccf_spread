```python
import seaborn.objects as so
from seaborn import load_dataset
from seaborn import axes_style
import pandas as pd
import numpy as np 
import matplotlib.offsetbox as offsetbox
```

```python
path = '/data/keeling/a/rytam2/ccf_model_spread/notebooks/finalized/globalfeedback/'
df = pd.read_csv(path+'df.csv') #model kernels * model dCCF
df_ceres = pd.read_csv(path+'df_ceres.csv') #ceres kernels * model CCF
df_modis = pd.read_csv(path+'df_modis.csv') #modis kernels * model CCF
df_isccp = pd.read_csv(path+'df_isccp.csv') #isccp kernels * model CCF
df_patmosx = pd.read_csv(path+'df_patmosx.csv') #patmosx kernels * model CCF
df_ccfavg = pd.read_csv(path+'df_ccfavg.csv') #model kernels * model CCF average
df_kernelavg = pd.read_csv(path+'df_kernelavg.csv')
```

```python
def make_textbox(axes, string):

    box1 = offsetbox.TextArea(string,textprops=dict(fontsize=12,ha='left',fontweight='bold'))
    anchored_box = offsetbox.AnchoredOffsetbox(loc=3,
                                 child=box1, pad=0.2,
                                 frameon=False,
                                 bbox_to_anchor=(0,1),
                                 #bbox_transform=axes.transAxes,
                                 borderpad=.2)
    axes.add_artist(anchored_box)
    
    return
```

## dR_hat_models * models kernel fig

```python
p1=so.Plot(data=df, x="var", y="dR/dT", color="experiment")\
                  .theme({**axes_style("whitegrid")})\
                  .label(color=str.capitalize,marker=str.capitalize,x="",y="")\
                  .scale(color=["#E69F00", "#0072B2", "#CC79A7","#009E73"],pointsize=20,y=so.Continuous().tick(every=0.5),)\
                  .add(so.Dot(pointsize=18), so.Dodge(gap=.5))\
                  .add(so.Range(color='k',alpha=.8), so.Est(errorbar="sd"), so.Dodge())\
                  .add(so.Dot(color='red',marker='_',pointsize=22,stroke=2), so.Est(func='mean'), so.Dodge())\
                  .limit(y=(-1, 1))\
                  .label(title="a) $\partial  R / \partial  CCF_{model}$ *$\hat{dCCF_{model}}$ ")\
                  .save('/data/keeling/a/rytam2/ccf_model_spread/notebooks/figures/comp/dotplot_dRg_allexperiments.png',  dpi=400, bbox_inches='tight')
```
